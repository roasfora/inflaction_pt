Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/inflacao_yoy_com_grupos_2019_2025.csv')
```

```python
df["Month"] = pd.to_datetime(df["Month"])
df.dtypes
```

# Gráfico: comparação por grupo

```python
plt.figure(figsize=(12,6))

sns.lineplot(data=df, x="Month", y="Total", label ="Total")
sns.lineplot(data=df, x="Month", y="Excluindo Alimentos e Energia", label ="Excluindo Alimentos e Energia")
sns.lineplot(data=df, x="Month", y="Bens", label ="Bens")
sns.lineplot(data=df, x="Month", y="Serviços", label ="Serviços")

plt.title("Inflação YoY por Grupo")
plt.xlabel("Mês")
plt.ylabel("Inflação YoY")
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

# Heatmap de correlação

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.drop(columns="Month").corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre Grupos de Inflação YoY")
plt.tight_layout()
plt.show()
```
